```python
url="https://raw.githubusercontent.com/adiacla/bigdata/master/casasOrg.csv"
df=pd.read_csv(url,sep=";")
#sep para separar por comas
df
```

```python
df.age=df.age+3
df
```

```python
#poner los nombre de las columnas
df.columns=["precio","metros_totales", "antiguedad", "precio_terreno", "metros_habitables","universitarios",
            "dormitorios", "chimenea", "banyos", "habitaciones","calefaccion","consumo_calefacion",
            "desague", "vistas_lago", "nueva_construccion","aire_acondicionado","fecha construido","numero catastro","MetroPorSalon","categoria"]
```

```python
dfc= df.copy()
```

```python
dfc
```

```python
dfc.drop("fecha construido",axis=1,inplace=True) # Borrar Columna  #axis1 = columnas #axis 0 = fila
dfc
```

```python
df.antiguedad.plot.hist()  # graficas en histograma
```

```python
dfc[dfc["antiguedad"]>1000] # Filtrar los datos con antiguedad mil años
```

```python
dfc.antiguedad.value_counts().sort_index(ascending=False) #misma funcion anerior para una sola palabra # sort para ordenarlos
```

```python
dfc.antiguedad=np.where((dfc.antiguedad<1000),dfc.antiguedad,(dfc.antiguedad-3/1000)+3) # Valores de antiguedad menores de mil
dfc.head(10)
```

```python
dfc.antiguedad=dfc.antiguedad.astype("int32")
dfc.info()
```

```python
dfc.isnull().sum()
```

```python
dfc.chimenea.fillna(0,inplace=True)
dfc.desague.fillna(method="ffill",inplace= True) # Remplazar filas deepues
```
